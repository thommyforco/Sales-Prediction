```python
"""
╔══════════════════════════════════════════════════════════════════════════════╗
║   MDFE v2 — Multimodal Demand Forecasting Engine                           ║
║   Dataset: Historical Product Demand (felixzhao/Kaggle)                    ║
║   Grossista multi-magazzino | 4+ anni di storico | LightGBM + Croston      ║
╠══════════════════════════════════════════════════════════════════════════════╣
║  Colonne CSV attese:                                                        ║
║    Product_Code, Warehouse, Product_Category, Date, Order_Demand           ║
╚══════════════════════════════════════════════════════════════════════════════╝

Utilizzo in Jupyter:
    engine = MDFEngine("Historical Product Demand.csv")
    engine.run()
    preds = engine.predict(horizon=4)
    engine.save("previsioni_wholesale.csv")
"""

import warnings
import logging
import sys
import re
import numpy as np
import pandas as pd
from pathlib import Path
from typing import Optional, Dict, Tuple

import lightgbm as lgb
from sklearn.metrics import mean_absolute_error, mean_squared_error

warnings.filterwarnings("ignore")
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    datefmt="%H:%M:%S",
)
log = logging.getLogger("MDFE")


# ──────────────────────────────────────────────────────────────────────────────
# CONFIGURAZIONE
# ──────────────────────────────────────────────────────────────────────────────
class Config:
    # Colonne CSV originali
    RAW_PRODUCT   = "Product_Code"
    RAW_WAREHOUSE = "Warehouse"
    RAW_CATEGORY  = "Product_Category"
    RAW_DATE      = "Date"
    RAW_DEMAND    = "Order_Demand"

    # Colonne interne
    COL_ID      = "ID"          # Product_Code + "_" + Warehouse
    COL_DATE    = "Data"
    COL_TARGET  = "Vendite"
    COL_PRICE   = "Peso"        # Proxy economico da categoria (no prezzo reale)

    # Lag — con 4+ anni Lag_52 è valido
    LAGS         = [1, 2, 4, 8, 13, 26, 52]

    # Rolling
    ROLLING_WINS = [4, 12, 26]

    # ABC
    ABC_A_THRESH = 0.80
    ABC_B_THRESH = 0.95

    # Outlier clip
    CLIP_PERCENTILE = 99

    # Storico minimo (settimane con vendite > 0)
    MIN_WEEKS_HISTORY = 12

    # Bayesian shrinkage per STE (con 4+ anni funziona)
    STE_SHRINKAGE_K = 3

    # LightGBM base
    LGBM_BASE = dict(
        n_estimators      = 5000,
        learning_rate     = 0.02,
        num_leaves        = 63,
        min_child_samples = 20,
        subsample         = 0.8,
        colsample_bytree  = 0.8,
        reg_alpha         = 0.1,
        reg_lambda        = 0.2,
        random_state      = 42,
        n_jobs            = -1,
        verbose           = -1,
    )

    LGBM_A = {**LGBM_BASE,
        "objective":          "huber",
        "alpha":               0.9,
        "learning_rate":       0.015,
        "num_leaves":          63,    # Ridotto: con 123 serie, 127 foglie = overfitting
        "min_child_samples":   30,    # Aumentato: forza split su pattern più stabili
        "reg_alpha":           0.3,   # L1 più forte per classe A (poche serie)
        "reg_lambda":          0.5,   # L2 più forte
    }
    LGBM_B = {**LGBM_BASE, "objective": "regression_l1"}
    LGBM_C = {**LGBM_BASE, "objective": "tweedie", "tweedie_variance_power": 1.5}

    EARLY_STOPPING_A = 150   # Più rounds per trovare il vero minimo
    EARLY_STOPPING   = 100    # Aumentato anche per B
    VALID_WEEKS_FIXED = 16   # ~4 mesi di validation — distribuiti su più stagioni


# ──────────────────────────────────────────────────────────────────────────────
# STEP 0 — LOADER WHOLESALE
# ──────────────────────────────────────────────────────────────────────────────
class WholesaleLoader:
    """
    Gestisce le specificità del dataset felixzhao:
    - Date in formato YYYY/M/D
    - Order_Demand con parentesi per negativi: (500) = -500
    - Nessuna colonna prezzo → peso proxy da categoria
    - ID = Product_Code + Warehouse (stesso prodotto in magazzini diversi = serie separate)
    """

    def __init__(self, path: str):
        self.path = Path(path)

    def load_and_aggregate(self) -> pd.DataFrame:
        log.info(f"Caricamento: {self.path}")

        df = pd.read_csv(self.path, dtype=str)
        log.info(f"Righe grezze: {df.shape[0]:,}")

        # ── Pulizia Order_Demand ──────────────────────────────────────────────
        # Rimuovi spazi, converti (500) → -500
        df[Config.RAW_DEMAND] = (
            df[Config.RAW_DEMAND]
            .str.strip()
            .str.replace(r"^\((.+)\)$", r"-\1", regex=True)  # (500) → -500
        )
        df[Config.RAW_DEMAND] = pd.to_numeric(df[Config.RAW_DEMAND], errors="coerce")

        # ── Pulizia date (YYYY/M/D → datetime) ───────────────────────────────
        df[Config.RAW_DATE] = pd.to_datetime(
            df[Config.RAW_DATE], format="%Y/%m/%d", errors="coerce"
        )

        # Rimuovi righe con dati mancanti o domanda negativa/nulla
        before = len(df)
        df = df.dropna(subset=[Config.RAW_DATE, Config.RAW_DEMAND, Config.RAW_PRODUCT])
        df = df[df[Config.RAW_DEMAND] > 0]
        log.info(f"Righe dopo pulizia: {len(df):,} (rimosse {before - len(df):,})")

        # ── ID combinato: prodotto + magazzino ────────────────────────────────
        df[Config.COL_ID] = df[Config.RAW_PRODUCT].str.strip() + "_" + df[Config.RAW_WAREHOUSE].str.strip()

        # ── Aggregazione settimanale ──────────────────────────────────────────
        df["Week"] = df[Config.RAW_DATE].dt.to_period("W").dt.start_time
        weekly = (
            df.groupby([Config.COL_ID, "Week", Config.RAW_CATEGORY])[Config.RAW_DEMAND]
            .sum()
            .reset_index()
            .rename(columns={
                "Week":              Config.COL_DATE,
                Config.RAW_DEMAND:   Config.COL_TARGET,
            })
        )

        # Mantieni categoria per ogni ID (usa la più frequente)
        cat_map = (
            df.groupby(Config.COL_ID)[Config.RAW_CATEGORY]
            .agg(lambda x: x.mode().iloc[0])
            .to_dict()
        )
        weekly[Config.RAW_CATEGORY] = weekly[Config.COL_ID].map(cat_map)

        # ── Peso proxy da categoria ───────────────────────────────────────────
        # Senza prezzo reale, usiamo la domanda mediana per categoria come proxy
        # (categorie con domanda più alta tendono ad avere più valore economico)
        cat_weight = (
            weekly.groupby(Config.RAW_CATEGORY)[Config.COL_TARGET]
            .median()
            .clip(lower=1)
        )
        # Normalizza log-scale [1, 10]
        lw = np.log1p(cat_weight)
        cat_weight_norm = 1 + 9 * (lw - lw.min()) / (lw.max() - lw.min() + 1e-6)
        weekly[Config.COL_PRICE] = weekly[Config.RAW_CATEGORY].map(cat_weight_norm.to_dict()).fillna(1.0)

        # ── Griglia completa (settimane mancanti → 0) ─────────────────────────
        all_ids   = weekly[Config.COL_ID].unique()
        all_weeks = pd.date_range(
            weekly[Config.COL_DATE].min(),
            weekly[Config.COL_DATE].max(),
            freq="W-MON",
        )
        log.info(
            f"Griglia: {len(all_ids):,} serie × {len(all_weeks)} settimane = "
            f"{len(all_ids) * len(all_weeks):,} righe potenziali"
        )

        idx = pd.MultiIndex.from_product(
            [all_ids, all_weeks], names=[Config.COL_ID, Config.COL_DATE]
        )
        weekly_full = (
            weekly.set_index([Config.COL_ID, Config.COL_DATE])[[Config.COL_TARGET]]
            .reindex(idx, fill_value=0)
            .reset_index()
        )

        # Re-aggiungi categoria e peso
        weekly_full[Config.RAW_CATEGORY] = weekly_full[Config.COL_ID].map(cat_map)
        weekly_full[Config.COL_PRICE]    = weekly_full[Config.COL_ID].map(
            weekly.groupby(Config.COL_ID)[Config.COL_PRICE].first().to_dict()
        ).fillna(1.0)

        # ── Filtra serie con storico insufficiente ────────────────────────────
        weeks_with_sales = (
            weekly_full[weekly_full[Config.COL_TARGET] > 0]
            .groupby(Config.COL_ID)[Config.COL_DATE]
            .count()
        )
        valid = weeks_with_sales[weeks_with_sales >= Config.MIN_WEEKS_HISTORY].index
        weekly_full = weekly_full[weekly_full[Config.COL_ID].isin(valid)]

        log.info(
            f"Serie valide (≥{Config.MIN_WEEKS_HISTORY} settimane): {len(valid):,}"
        )
        log.info(
            f"Dataset aggregato: {len(weekly_full):,} righe | "
            f"periodo: {weekly_full[Config.COL_DATE].min().date()} → "
            f"{weekly_full[Config.COL_DATE].max().date()}"
        )
        return weekly_full.sort_values([Config.COL_ID, Config.COL_DATE]).reset_index(drop=True)


# ──────────────────────────────────────────────────────────────────────────────
# STEP 1 — SEGMENTAZIONE ABC
# ──────────────────────────────────────────────────────────────────────────────
class ABCSegmenter:

    def __init__(self):
        self.mapping_: Dict = {}

    def fit(self, df: pd.DataFrame) -> "ABCSegmenter":
        revenue = (
            df.groupby(Config.COL_ID)
            .apply(lambda x: (x[Config.COL_TARGET] * x[Config.COL_PRICE]).sum())
            .sort_values(ascending=False)
            .reset_index()
        )
        revenue.columns = [Config.COL_ID, "Valore"]
        total = revenue["Valore"].sum()
        revenue["Cum_Pct"] = revenue["Valore"].cumsum() / total

        def _cls(p):
            if p <= Config.ABC_A_THRESH: return "A"
            if p <= Config.ABC_B_THRESH: return "B"
            return "C"

        revenue["Classe"] = revenue["Cum_Pct"].apply(_cls)
        self.mapping_ = dict(zip(revenue[Config.COL_ID], revenue["Classe"]))

        for c in ["A", "B", "C"]:
            n   = sum(v == c for v in self.mapping_.values())
            pct = n / len(self.mapping_) * 100
            log.info(f"  Classe {c}: {n:,} serie ({pct:.1f}%)")
        return self

    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy()
        df["Classe_ABC"] = df[Config.COL_ID].map(self.mapping_).fillna("C")
        return df


# ──────────────────────────────────────────────────────────────────────────────
# STEP 2 — PRE-PROCESSING
# ──────────────────────────────────────────────────────────────────────────────
class PreProcessor:

    def __init__(self):
        self.clip_threshold_: Optional[float] = None

    def fit(self, df: pd.DataFrame) -> "PreProcessor":
        self.clip_threshold_ = np.percentile(
            df[Config.COL_TARGET].dropna(), Config.CLIP_PERCENTILE
        )
        log.info(
            f"Outlier clip al {Config.CLIP_PERCENTILE}° percentile: "
            f"{self.clip_threshold_:.1f} unità/settimana"
        )
        return self

    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy()
        df[Config.COL_TARGET] = df[Config.COL_TARGET].clip(upper=self.clip_threshold_)
        df["Target_Log"]      = np.log1p(df[Config.COL_TARGET])
        df = self._calendar(df)
        return df

    @staticmethod
    def _calendar(df: pd.DataFrame) -> pd.DataFrame:
        d = df[Config.COL_DATE]
        df["Settimana_Anno"]   = d.dt.isocalendar().week.astype(int)
        df["Mese"]             = d.dt.month
        df["Trimestre"]        = d.dt.quarter
        df["Anno"]             = d.dt.year
        df["Giorno_Settimana"] = d.dt.dayofweek

        # Ciclica
        df["Sin_Settimana"] = np.sin(2 * np.pi * df["Settimana_Anno"] / 52)
        df["Cos_Settimana"] = np.cos(2 * np.pi * df["Settimana_Anno"] / 52)
        df["Sin_Mese"]      = np.sin(2 * np.pi * df["Mese"] / 12)
        df["Cos_Mese"]      = np.cos(2 * np.pi * df["Mese"] / 12)

        # Flag stagionali wholesale
        df["Is_Q4"]        = (df["Trimestre"] == 4).astype(int)
        df["Is_Q1"]        = (df["Trimestre"] == 1).astype(int)  # post-feste, calo ordini
        df["Is_Estate"]    = df["Mese"].isin([7, 8]).astype(int)
        df["Fine_Anno"]    = (df["Settimana_Anno"] >= 48).astype(int)
        df["Inizio_Anno"]  = (df["Settimana_Anno"] <= 4).astype(int)
        return df

    @staticmethod
    def inverse(y_log: np.ndarray) -> np.ndarray:
        return np.expm1(np.maximum(y_log, 0))


# ──────────────────────────────────────────────────────────────────────────────
# STEP 3 — FEATURE ENGINEERING (con STE attivato — 4+ anni di storico)
# ──────────────────────────────────────────────────────────────────────────────
class FeatureEngineer:

    def __init__(self):
        self.te_map_        : Dict = {}
        self.seasonal_te_   : Dict = {}
        self.cat_week_map_  : Dict = {}   # Feature aggregate per categoria
        self.id_cat_map_    : Dict = {}   # Mappa ID → categoria
        self._global_mean   : float = 0.0

    def fit_target_encoding(self, df_train: pd.DataFrame) -> "FeatureEngineer":
        # Target encoding globale
        self.te_map_      = df_train.groupby(Config.COL_ID)[Config.COL_TARGET].mean().to_dict()
        self._global_mean = df_train[Config.COL_TARGET].mean()

        # ── Seasonal TE con Bayesian Shrinkage ────────────────────────────────
        # Con 4+ anni abbiamo 4 osservazioni per settimana → STE affidabile
        k = Config.STE_SHRINKAGE_K
        stats = (
            df_train.groupby([Config.COL_ID, "Settimana_Anno"])[Config.COL_TARGET]
            .agg(["mean", "count"])
            .reset_index()
            .rename(columns={"mean": "STE_Raw", "count": "N"})
        )
        stats["Media_ID"] = stats[Config.COL_ID].map(self.te_map_).fillna(self._global_mean)
        stats["STE_Idx"]  = stats["STE_Raw"] / (stats["Media_ID"] + 1e-6)
        # Smoothing: poche obs → avvicina a 1.0, molte obs → usa dato reale
        stats["STE"]      = (stats["N"] * stats["STE_Idx"] + k) / (stats["N"] + k)
        self.seasonal_te_ = (
            stats.set_index([Config.COL_ID, "Settimana_Anno"])["STE"].to_dict()
        )
        # ── Feature aggregate per categoria ──────────────────────────────────
        # Con 123 serie in classe A, il modello ha pochi esempi.
        # Aggiungere la domanda media della categoria per settimana
        # permette il cross-learning tra prodotti simili.
        if Config.RAW_CATEGORY in df_train.columns:
            self.id_cat_map_ = df_train.groupby(Config.COL_ID)[Config.RAW_CATEGORY].first().to_dict()
            cat_week = (
                df_train.groupby([Config.RAW_CATEGORY, "Settimana_Anno"])[Config.COL_TARGET]
                .agg(["mean", "sum"])
                .reset_index()
                .rename(columns={"mean": "Cat_Week_Mean", "sum": "Cat_Week_Sum"})
            )
            self.cat_week_map_ = {
                (r[Config.RAW_CATEGORY], r["Settimana_Anno"]): (r["Cat_Week_Mean"], r["Cat_Week_Sum"])
                for _, r in cat_week.iterrows()
            }
            log.info(f"Category features: {len(self.cat_week_map_):,} coppie (categoria × settimana)")

        log.info(
            f"Target encoding: {len(self.te_map_):,} serie | "
            f"Seasonal TE: {len(self.seasonal_te_):,} coppie (k={k})"
        )
        return self

    def transform(self, df: pd.DataFrame) -> pd.DataFrame:
        df = df.copy().sort_values([Config.COL_ID, Config.COL_DATE])
        grp = df.groupby(Config.COL_ID)[Config.COL_TARGET]

        # Lag strategici — Lag_52 ora valido con 4+ anni
        for lag in Config.LAGS:
            df[f"Lag_{lag}"] = grp.shift(lag)

        # Rolling stats
        for win in Config.ROLLING_WINS:
            roll = grp.shift(1).rolling(win, min_periods=1)
            df[f"Roll_Mean_{win}"] = roll.mean()
            df[f"Roll_Std_{win}"]  = roll.std().fillna(0)
            df[f"CoV_{win}"]       = df[f"Roll_Std_{win}"] / (df[f"Roll_Mean_{win}"] + 1e-6)

        # Momentum e trend
        df["Momentum_4w"]    = df["Lag_1"] - df["Roll_Mean_4"]
        df["Momentum_12w"]   = df["Lag_1"] - df["Roll_Mean_12"]
        df["Rate_Of_Change"] = (df["Lag_1"] - df["Lag_2"]) / (df["Lag_2"] + 1e-6)
        df["Trend_4_12"]     = df["Roll_Mean_4"] - df["Roll_Mean_12"]
        df["Trend_12_26"]    = df["Roll_Mean_12"] - df["Roll_Mean_26"]

        # YoY ratio: questa settimana vs stessa settimana anno scorso
        df["YoY_Ratio"] = df["Lag_1"] / (df["Lag_52"] + 1e-6)
        df["YoY_Ratio"] = df["YoY_Ratio"].clip(0, 5)  # Cap a 5x per stabilità

        # Target encoding
        df["TE_ID"] = df[Config.COL_ID].map(self.te_map_).fillna(self._global_mean)

        # Seasonal TE (ora affidabile con 4+ anni)
        df["STE"] = df.apply(
            lambda r: self.seasonal_te_.get((r[Config.COL_ID], r["Settimana_Anno"]), 1.0),
            axis=1,
        )
        df["STE_Assoluto"] = df["STE"] * df["TE_ID"]

        # ABC numerico
        df["ABC_Num"] = df["Classe_ABC"].map({"A": 2, "B": 1, "C": 0}).fillna(0)

        # ── Feature aggregate per categoria ──────────────────────────────────
        # Media e volume totale della categoria in quella settimana dell'anno
        # Permette cross-learning tra prodotti della stessa famiglia
        if self.cat_week_map_:
            df["Categoria"] = df[Config.COL_ID].map(self.id_cat_map_)
            df["Cat_Week_Mean"] = df.apply(
                lambda r: self.cat_week_map_.get((r.get("Categoria"), r["Settimana_Anno"]), (0, 0))[0],
                axis=1,
            )
            df["Cat_Week_Sum"] = df.apply(
                lambda r: self.cat_week_map_.get((r.get("Categoria"), r["Settimana_Anno"]), (0, 0))[1],
                axis=1,
            )
            # Share storica del prodotto sulla categoria
            # IMPORTANTE: usa Lag_1 (non il target corrente) per evitare data leakage
            lag1 = df.groupby(Config.COL_ID)[Config.COL_TARGET].shift(1).fillna(0)
            df["Cat_Share"] = lag1 / (df["Cat_Week_Sum"] + 1e-6)
            df["Cat_Share"] = df["Cat_Share"].clip(0, 1)
        else:
            df["Cat_Week_Mean"] = 0.0
            df["Cat_Week_Sum"]  = 0.0
            df["Cat_Share"]     = 0.0

        return df

    @staticmethod
    def feature_cols() -> list:
        cols = [f"Lag_{l}" for l in Config.LAGS]
        for w in Config.ROLLING_WINS:
            cols += [f"Roll_Mean_{w}", f"Roll_Std_{w}", f"CoV_{w}"]
        cols += [
            "Momentum_4w", "Momentum_12w", "Rate_Of_Change",
            "Trend_4_12", "Trend_12_26",
            "YoY_Ratio",
            "TE_ID", "STE", "STE_Assoluto", "ABC_Num",
            "Settimana_Anno", "Mese", "Trimestre", "Anno",
            "Sin_Settimana", "Cos_Settimana", "Sin_Mese", "Cos_Mese",
            "Is_Q4", "Is_Q1", "Is_Estate", "Fine_Anno", "Inizio_Anno",
            "Cat_Week_Mean", "Cat_Week_Sum", "Cat_Share",
        ]
        return cols


# ──────────────────────────────────────────────────────────────────────────────
# STEP 4 — ECONOMIC WEIGHTS
# ──────────────────────────────────────────────────────────────────────────────
def economic_weights(df: pd.DataFrame) -> np.ndarray:
    p = df[Config.COL_PRICE].fillna(1.0).clip(lower=1e-6)
    lp = np.log1p(p)
    w_min, w_max = lp.min(), lp.max()
    if w_max > w_min:
        return (1 + 9 * (lp - w_min) / (w_max - w_min)).values
    return np.ones(len(p))


# ──────────────────────────────────────────────────────────────────────────────
# STEP 5 — CROSTON TSB (Classe C)
# ──────────────────────────────────────────────────────────────────────────────
class CrostonForecaster:

    def __init__(self, alpha: float = 0.1, beta: float = 0.1):
        self.alpha = alpha
        self.beta  = beta
        self.params_: Dict = {}

    def fit(self, df: pd.DataFrame) -> "CrostonForecaster":
        for pid, grp in df.groupby(Config.COL_ID):
            series = grp.sort_values(Config.COL_DATE)[Config.COL_TARGET].values
            self.params_[pid] = self._fit_series(series)
        log.info(f"  Croston TSB: fittate {len(self.params_):,} serie Classe C")
        return self

    def _fit_series(self, series: np.ndarray) -> tuple:
        alpha, beta = self.alpha, self.beta
        nz = series[series > 0]
        z  = nz.mean() if len(nz) > 0 else 1.0
        p  = len(nz) / len(series) if len(series) > 0 else 0.1
        for y in series:
            if y > 0:
                z = alpha * y + (1 - alpha) * z
                p = beta  * 1 + (1 - beta)  * p
            else:
                p = beta  * 0 + (1 - beta)  * p
        return z, p

    def predict(self, product_ids: np.ndarray) -> np.ndarray:
        return np.array([
            max(self.params_[pid][0] * self.params_[pid][1], 0.0)
            if pid in self.params_ else 0.0
            for pid in product_ids
        ])


# ──────────────────────────────────────────────────────────────────────────────
# STEP 6 — TRAINER
# ──────────────────────────────────────────────────────────────────────────────
class ClassTrainer:
    _CFGS = {"A": Config.LGBM_A, "B": Config.LGBM_B, "C": Config.LGBM_C}

    def __init__(self, cls: str):
        self.cls   = cls
        self.model_: Optional[lgb.LGBMRegressor] = None
        self.fi_   : Optional[pd.Series]         = None

    def fit(self, X_tr, y_tr, X_va, y_va, weights=None):
        es = Config.EARLY_STOPPING_A if self.cls == "A" else Config.EARLY_STOPPING
        self.model_ = lgb.LGBMRegressor(**self._CFGS[self.cls])
        self.model_.fit(
            X_tr, y_tr,
            eval_set=[(X_va, y_va)],
            sample_weight=weights,
            callbacks=[
                lgb.early_stopping(es, verbose=False),
                lgb.log_evaluation(period=-1),
            ],
        )
        self.fi_ = pd.Series(
            self.model_.feature_importances_, index=X_tr.columns
        ).sort_values(ascending=False)
        log.info(
            f"  Classe {self.cls} → best_iter={self.model_.best_iteration_} | "
            f"top feature: {self.fi_.index[0]}"
        )
        return self

    def predict(self, X: pd.DataFrame) -> np.ndarray:
        return self.model_.predict(X)


# ──────────────────────────────────────────────────────────────────────────────
# STEP 7 — EVALUATOR + BENCHMARK
# ──────────────────────────────────────────────────────────────────────────────
class Evaluator:
    @staticmethod
    def metrics(y_true, y_pred, prices):
        mae   = mean_absolute_error(y_true, y_pred)
        rmse  = np.sqrt(mean_squared_error(y_true, y_pred))
        mask  = y_true >= 5
        mape  = np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100 \
                if mask.any() else 0.0
        wmape = np.sum(np.abs(y_true - y_pred)) / (np.sum(y_true) + 1e-6) * 100
        eco   = np.mean(np.abs(y_true - y_pred) * prices)
        return {
            "MAE":             round(mae,   2),
            "RMSE":            round(rmse,  2),
            "MAPE_% (qty>=5)": round(mape,  1),
            "wMAPE_% (KEY)":   round(wmape, 1),
            "Economic_Error":  round(eco,   2),
        }

    @staticmethod
    def report(results):
        print("\n" + "═" * 56)
        print("  MDFE v2 — PERFORMANCE REPORT")
        print("═" * 56)
        for cls, m in results.items():
            print(f"\n  Classe {cls}:")
            for k, v in m.items():
                print(f"    {k:<24} {v:>10}")
        print("═" * 56 + "\n")


class NaiveBenchmark:
    def fit_predict(self, df_train, df_val):
        results = []
        for pid in df_val[Config.COL_ID].unique():
            hist = (
                df_train[df_train[Config.COL_ID] == pid]
                .sort_values(Config.COL_DATE)[Config.COL_TARGET].values
            )
            val_rows = df_val[df_val[Config.COL_ID] == pid].sort_values(Config.COL_DATE)
            if len(hist) == 0:
                continue
            for _, row in val_rows.iterrows():
                naive    = hist[-1]
                ma4      = hist[-4:].mean()  if len(hist) >= 4  else hist.mean()
                ma12     = hist[-12:].mean() if len(hist) >= 12 else hist.mean()
                seasonal = hist[-52]         if len(hist) >= 52 else hist.mean()
                results.append({
                    Config.COL_ID:    pid,
                    Config.COL_DATE:  row[Config.COL_DATE],
                    "Classe_ABC":     row["Classe_ABC"],
                    Config.COL_PRICE: row[Config.COL_PRICE],
                    "y_true":         row[Config.COL_TARGET],
                    "Naive":          max(naive,    0),
                    "MA4":            max(ma4,      0),
                    "MA12":           max(ma12,     0),
                    "Seasonal_Naive": max(seasonal, 0),
                })
                hist = np.append(hist, row[Config.COL_TARGET])
        return pd.DataFrame(results)

    @staticmethod
    def wmape(y_true, y_pred):
        return float(np.sum(np.abs(y_true - y_pred)) / (np.sum(y_true) + 1e-6) * 100)

    @staticmethod
    def eco(y_true, y_pred, prices):
        return float(np.mean(np.abs(y_true - y_pred) * prices))

    def report(self, df, mdfe_results):
        models = ["Naive", "MA4", "MA12", "Seasonal_Naive"]
        print("\n" + "═" * 76)
        print("  CONFRONTO MDFE v2 vs BASELINE")
        print("═" * 76)
        print(f"  {'Cls':<5} {'Modello':<20} {'wMAPE%':>8} {'EcoErr':>9}  {'Δ wMAPE':>9}")
        print("─" * 76)
        for cls in ["A", "B", "C"]:
            sub = df[df["Classe_ABC"] == cls]
            if len(sub) == 0:
                continue
            y_true = sub["y_true"].values
            prices = sub[Config.COL_PRICE].values
            mdfe_w = mdfe_results.get(cls, {}).get("wMAPE_% (KEY)")
            mdfe_e = mdfe_results.get(cls, {}).get("Economic_Error")
            if mdfe_w is not None:
                print(f"  {cls:<5} {'MDFE v2':<20} {mdfe_w:>7.1f}% {mdfe_e:>9.2f}  {'← nostro'}")
            for m in models:
                w = self.wmape(y_true, sub[m].values)
                e = self.eco(y_true, sub[m].values, prices)
                delta = f"{mdfe_w - w:+.1f}%" if mdfe_w else "N/A"
                icon  = "✅" if mdfe_w and mdfe_w < w else "🔴"
                print(f"  {cls:<5} {m:<20} {w:>7.1f}% {e:>9.2f}  {delta:>8} {icon}")
            print()
        print("═" * 76)
        print("  ✅ MDFE migliore  |  🔴 baseline migliore")
        print("═" * 76)


# ──────────────────────────────────────────────────────────────────────────────
# ORCHESTRATORE
# ──────────────────────────────────────────────────────────────────────────────
class MDFEngine:

    def __init__(self, data_path: str):
        self.data_path     = data_path
        self.df_agg_       : Optional[pd.DataFrame]  = None
        self.df_feat_      : Optional[pd.DataFrame]  = None
        self.trainers_     : Dict[str, ClassTrainer] = {}
        self.croston_      = CrostonForecaster()
        self.prep_         = PreProcessor()
        self.feat_eng_     = FeatureEngineer()
        self.abc_          = ABCSegmenter()
        self.preds_        : Optional[pd.DataFrame]  = None
        self.benchmark_df_ : Optional[pd.DataFrame]  = None
        self._fcols        = FeatureEngineer.feature_cols()

    # ── Training ──────────────────────────────────────────────────────────────
    def run(self):
        log.info("═" * 56)
        log.info("  AVVIO MDFE v2 — WHOLESALE PIPELINE")
        log.info("═" * 56)

        loader       = WholesaleLoader(self.data_path)
        self.df_agg_ = loader.load_and_aggregate()

        log.info("Segmentazione ABC…")
        self.abc_.fit(self.df_agg_)
        df = self.abc_.transform(self.df_agg_)

        df, df_train, df_val = self._split(df)

        log.info("Pre-processing…")
        self.prep_.fit(df_train)
        df_train = self.prep_.transform(df_train)
        df_val   = self.prep_.transform(df_val)

        log.info("Feature engineering…")
        self.feat_eng_.fit_target_encoding(df_train)
        df_train = self.feat_eng_.transform(df_train)
        df_val   = self.feat_eng_.transform(df_val)

        # Salva df completo per predict()
        df_all = self.feat_eng_.transform(
            self.prep_.transform(self.abc_.transform(self.df_agg_))
        )
        self.df_feat_ = df_all

        log.info("Addestramento modelli…")
        eval_results = {}

        for cls in ["A", "B", "C"]:
            mask_tr = df_train["Classe_ABC"] == cls
            mask_va = df_val["Classe_ABC"]   == cls

            if mask_tr.sum() < 50:
                log.warning(f"  Classe {cls}: campioni insufficienti, skip.")
                continue

            X_tr = df_train.loc[mask_tr, self._fcols].fillna(0)
            y_tr = df_train.loc[mask_tr, "Target_Log"]
            X_va = df_val.loc[mask_va, self._fcols].fillna(0)
            y_va = df_val.loc[mask_va, "Target_Log"]

            if cls == "C":
                self.croston_.fit(df_train.loc[mask_tr])
                if len(X_va) > 0:
                    pids   = df_val.loc[mask_va, Config.COL_ID].values
                    y_pred = self.croston_.predict(pids)
                    y_true = df_val.loc[mask_va, Config.COL_TARGET].values
                    prices = df_val.loc[mask_va, Config.COL_PRICE].values
                    eval_results[cls] = Evaluator.metrics(y_true, y_pred, prices)
            else:
                w = economic_weights(df_train.loc[mask_tr])
                t = ClassTrainer(cls).fit(X_tr, y_tr, X_va, y_va, weights=w)
                self.trainers_[cls] = t
                if len(X_va) > 0:
                    y_pred  = self.prep_.inverse(t.predict(X_va))
                    y_true  = self.prep_.inverse(y_va.values)
                    prices  = df_val.loc[mask_va, Config.COL_PRICE].values
                    eval_results[cls] = Evaluator.metrics(y_true, y_pred, prices)

        Evaluator.report(eval_results)

        log.info("Benchmark naive per confronto…")
        bench            = NaiveBenchmark()
        self.benchmark_df_ = bench.fit_predict(df_train, df_val)
        bench.report(self.benchmark_df_, eval_results)

        log.info("Training completato.")

    # ── Previsione futura ─────────────────────────────────────────────────────
    def predict(self, horizon: int = 4) -> pd.DataFrame:
        assert self.df_feat_ is not None, "Esegui .run() prima di .predict()"
        log.info(f"Generazione previsioni — orizzonte: {horizon} settimane")

        last_date = self.df_feat_[Config.COL_DATE].max()
        df_roll   = self.df_feat_.copy()
        all_preds = []

        for step in range(1, horizon + 1):
            next_date = last_date + pd.Timedelta(weeks=step)
            week_num  = next_date.isocalendar()[1]

            latest = (
                df_roll.sort_values(Config.COL_DATE)
                .groupby(Config.COL_ID).last().reset_index()
            )
            latest[Config.COL_DATE] = next_date
            latest = PreProcessor._calendar(latest)

            for lag in Config.LAGS:
                latest[f"Lag_{lag}"] = (
                    df_roll.sort_values(Config.COL_DATE)
                    .groupby(Config.COL_ID)[Config.COL_TARGET]
                    .nth(-lag).values
                )

            for win in Config.ROLLING_WINS:
                latest[f"Roll_Mean_{win}"] = (
                    df_roll.sort_values(Config.COL_DATE)
                    .groupby(Config.COL_ID)[Config.COL_TARGET]
                    .apply(lambda x: x.tail(win).mean()).values
                )
                latest[f"Roll_Std_{win}"] = (
                    df_roll.sort_values(Config.COL_DATE)
                    .groupby(Config.COL_ID)[Config.COL_TARGET]
                    .apply(lambda x: x.tail(win).std() or 0).values
                )
                latest[f"CoV_{win}"] = (
                    latest[f"Roll_Std_{win}"] / (latest[f"Roll_Mean_{win}"] + 1e-6)
                )

            latest["Momentum_4w"]    = latest["Lag_1"] - latest["Roll_Mean_4"]
            latest["Momentum_12w"]   = latest["Lag_1"] - latest["Roll_Mean_12"]
            latest["Rate_Of_Change"] = (latest["Lag_1"] - latest["Lag_2"]) / (latest["Lag_2"] + 1e-6)
            latest["Trend_4_12"]     = latest["Roll_Mean_4"]  - latest["Roll_Mean_12"]
            latest["Trend_12_26"]    = latest["Roll_Mean_12"] - latest["Roll_Mean_26"]
            latest["YoY_Ratio"]      = (latest["Lag_1"] / (latest["Lag_52"] + 1e-6)).clip(0, 5)
            latest["TE_ID"]          = latest[Config.COL_ID].map(self.feat_eng_.te_map_).fillna(0)
            latest["STE"]            = latest[Config.COL_ID].apply(
                lambda pid: self.feat_eng_.seasonal_te_.get((pid, week_num), 1.0)
            )
            latest["STE_Assoluto"]   = latest["STE"] * latest["TE_ID"]
            latest["ABC_Num"]        = latest["Classe_ABC"].map({"A":2,"B":1,"C":0}).fillna(0)

            # Category features per la settimana futura
            if self.feat_eng_.cat_week_map_:
                latest["Categoria"] = latest[Config.COL_ID].map(self.feat_eng_.id_cat_map_)
                latest["Cat_Week_Mean"] = latest.apply(
                    lambda r: self.feat_eng_.cat_week_map_.get((r.get("Categoria"), week_num), (0,0))[0], axis=1
                )
                latest["Cat_Week_Sum"] = latest.apply(
                    lambda r: self.feat_eng_.cat_week_map_.get((r.get("Categoria"), week_num), (0,0))[1], axis=1
                )
                latest["Cat_Share"] = 0.0   # Non noto per settimane future
            else:
                latest["Cat_Week_Mean"] = 0.0
                latest["Cat_Week_Sum"]  = 0.0
                latest["Cat_Share"]     = 0.0

            step_preds = []

            # A e B → LightGBM
            for cls, trainer in self.trainers_.items():
                mask = latest["Classe_ABC"] == cls
                if mask.sum() == 0:
                    continue
                X      = latest.loc[mask, self._fcols].fillna(0)
                y_pred = self.prep_.inverse(trainer.predict(X))
                sub    = latest.loc[mask, [Config.COL_ID, "Classe_ABC", Config.COL_PRICE]].copy()
                sub[Config.COL_DATE]    = next_date
                sub["Previsione_Qty"]   = np.round(np.maximum(y_pred, 0), 1)
                sub["Previsione_Valore"]= np.round(sub["Previsione_Qty"] * sub[Config.COL_PRICE], 2)
                step_preds.append(sub)

            # C → Croston
            mask_c = latest["Classe_ABC"] == "C"
            if mask_c.sum() > 0:
                pids_c   = latest.loc[mask_c, Config.COL_ID].values
                y_pred_c = self.croston_.predict(pids_c)
                sub_c    = latest.loc[mask_c, [Config.COL_ID, "Classe_ABC", Config.COL_PRICE]].copy()
                sub_c[Config.COL_DATE]     = next_date
                sub_c["Previsione_Qty"]    = np.round(np.maximum(y_pred_c, 0), 1)
                sub_c["Previsione_Valore"] = np.round(sub_c["Previsione_Qty"] * sub_c[Config.COL_PRICE], 2)
                step_preds.append(sub_c)

            if step_preds:
                sdf = pd.concat(step_preds, ignore_index=True)
                all_preds.append(sdf)
                upd = sdf[[Config.COL_ID, Config.COL_DATE, "Previsione_Qty",
                            "Classe_ABC", Config.COL_PRICE]].rename(
                    columns={"Previsione_Qty": Config.COL_TARGET}
                )
                upd["Target_Log"] = np.log1p(upd[Config.COL_TARGET])
                df_roll = pd.concat([df_roll, upd], ignore_index=True)

        self.preds_ = pd.concat(all_preds, ignore_index=True).sort_values(
            [Config.COL_ID, Config.COL_DATE]
        )
        log.info(f"Previsioni generate: {len(self.preds_):,} righe.")
        return self.preds_

    def save(self, path: str = "previsioni_wholesale.csv"):
        assert self.preds_ is not None, "Esegui .predict() prima di .save()"
        self.preds_.to_csv(path, index=False)
        log.info(f"Salvato: {path}")

    def feature_importance(self, top_n: int = 15) -> pd.DataFrame:
        frames = []
        for cls, t in self.trainers_.items():
            fi = t.fi_.head(top_n).reset_index()
            fi.columns = ["Feature", "Importanza"]
            fi["Classe"] = cls
            frames.append(fi)
        return pd.concat(frames, ignore_index=True) if frames else pd.DataFrame()

    def _split(self, df):
        dates  = df[Config.COL_DATE].sort_values().unique()
        n_val  = min(Config.VALID_WEEKS_FIXED, len(dates) - 52)  # Lascia almeno 1 anno di train
        cutoff = dates[-n_val]
        train  = df[df[Config.COL_DATE] <  cutoff]
        val    = df[df[Config.COL_DATE] >= cutoff]
        log.info(
            f"Split → train fino a {train[Config.COL_DATE].max().date()} | "
            f"val: {val[Config.COL_DATE].min().date()} → "
            f"{val[Config.COL_DATE].max().date()} ({n_val} settimane)"
        )
        return df, train, val


# ──────────────────────────────────────────────────────────────────────────────
# ENTRY POINT
# ──────────────────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    # ── Imposta qui i tuoi parametri ──────────────────────────────────────────
    data_path = "Historical Product Demand.csv"   # <-- nome file Kaggle
    horizon   = 4                                  # <-- settimane da prevedere
    # ─────────────────────────────────────────────────────────────────────────

    engine = MDFEngine(data_path=data_path)
    engine.run()

    preds = engine.predict(horizon=horizon)
    engine.save("previsioni_wholesale.csv")

    print("\n── Anteprima previsioni ─────────────────────────────")
    print(preds.head(20).to_string(index=False))

    print("\n── Top Feature Importance ───────────────────────────")
    fi = engine.feature_importance(top_n=10)
    if not fi.empty:
        print(fi.to_string(index=False))
```
